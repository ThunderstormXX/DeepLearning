```python
def check_grad( func , grad , X ) :
    dX = X/1000 
    df = func(X+dX) - func(X-dX)
    gd = grad(X)
    print('f(x) = ' ,func(X) )
    print('df(x)= ', grad(X))
    print("COMPUTATIONALS :")
    print('df : ',df)
    print('sum(grad.T*dx): ',tc.sum(2*gd*dX))
    print('grad.T@dX: ',gd@dX)
    if tc.sum(df-tc.sum(2*gd*dX)) < 0.01 or tc.sum(df -gd@dX) < 0.01  :
        print("OK")
    else :
        print('NOT OK')    
```

```python
def func(X) :
    return tc.sqrt(tc.trace(X.T@X))
def grad(X) :
    return X / tc.sqrt(tc.trace(X.T@X))
X = tc.FloatTensor([[3,4]])
check_grad(func,grad,X)
```

```python
L = tc.FloatTensor([[1,0,0],[0,2,0],[0,0,7]])
def func(Q) :
    return Q.T@L@Q
def grad(Q) :
    return (Q.T@L - tc.matmul(func(Q).T,Q.T)).T
Q = tc.FloatTensor([[1,0,2],[1,2,2],[12,2,7]])
check_grad(func,grad,Q)
```
